```python
from datetime import datetime

import numpy as np
import pandas as pd

from sklearn import preprocessing 
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, DotProduct, WhiteKernel, ConstantKernel as C
from sklearn import linear_model

import matplotlib.pyplot as plt

FAMA_49CRSP = 'FAMA_49CRSP.csv'
```

```python
def encode(df, col):
    """
    Takes in pandas dataframe and encodes df.col as numbers. 
    """
    encoder = preprocessing.LabelEncoder()
    industry_desc_encode = encoder.fit_transform(df[col])
    df[col] = industry_desc_encode
    return df 

def process_data(filename):
    """
    Reads in a CSV file and encodes text columns. 
    Removes indret_ew, indret_vw from df. 
    Returns df (X matrix) and ew_indret (Y values )
    """
    df = pd.read_csv(filename, header=0, dtype={'public_date' : str})
    df = encode(df, 'FFI49_desc')
    
    ew_indret = df.indret_ew
    
    df = df.drop(labels=['indret_ew', 'indret_vw'], axis=1)
        
    return df, ew_indret

def split_data(x, y): 
    """
    Splits data into 0.64 Train, 0.16 dev, 0.2 Test
    """
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle = False)
    x_train, x_dev, y_train, y_dev = train_test_split(x_train, y_train, test_size = 0.2, shuffle = False)
    
    return x_train, x_test, x_dev, y_dev, y_train, y_test
```

```python
def split_industries():
    df = pd.read_csv(FAMA_49CRSP)
    
    industries = set(df['FFI49_desc'])
    
    for ind in industries: 
        df_ind = df[df['FFI49_desc'] == ind]
        df_ind.to_csv('industries/{}.csv'.format(ind))
```

```python
def dev(model, x_train, y_train, x_dev, y_dev):
    # also try RMSE? 
    y_train_predict = model.predict(x_train)
    train_mse = metrics.mean_squared_error(y_train, y_train_predict)
    
    y_dev_predict = model.predict(x_dev)
    mse = metrics.mean_squared_error(y_dev, y_dev_predict)
    r2 = metrics.r2_score(y_dev, y_dev_predict)
    
    print('R2: {}\tMSE: {}'.format(r2, mse))
    return y_train_predict, y_dev_predict
```

```python
df, ew_indret = process_data('industries/AERO.csv')
x_train, x_test, x_dev, y_dev, y_train, y_test = split_data(df, ew_indret)
```

```python
def train_GPR(x_train, y_train):
    kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
    gpr = GaussianProcessRegressor(kernel=kernel)
    gpr.fit(x_train, y_train)
    return gpr

def train_LinearRegression(x_train, y_train):
    regr = linear_model.LinearRegression()
    regr.fit(x_train, y_train)
    return regr

def train_ElasticNet(alpha, l1, x_train, y_train):
    reg = linear_model.ElasticNet(alpha=alpha, l1_ratio=l1)
    reg.fit(x_train, y_train) 
    return reg

def train_Ridge(alpha, x_train, y_train):
    reg = linear_model.Ridge(alpha=alpha)
    reg.fit(x_train, y_train) 
    return reg
```

```python
train_dates = [datetime(year=int(x[0:4]), month=int(x[4:6]), day=int(x[6:8])) for x in x_train['public_date']] 
dev_dates = [datetime(year=int(x[0:4]), month=int(x[4:6]), day=int(x[6:8])) for x in x_dev['public_date']]

x_train = x_train.drop('public_date', axis=1)
x_dev = x_dev.drop('public_date', axis=1)
```

```python
regr = train_LinearRegression(x_train, y_train)
regr_train_pred, regr_dev_pred = dev(regr, x_train, y_train, x_dev, y_dev)

graph(train_dates, y_train, regr_train_pred)
graph(dev_dates, y_dev, regr_dev_pred)
```

```python
reg = train_Ridge(0.1, x_train, y_train)
ridge_train_pred, ridge_dev_pred = dev(reg, x_train, y_train, x_dev, y_dev)

graph(train_dates, y_train, ridge_train_pred)
graph(dev_dates, y_dev, ridge_dev_pred)
```

```python
gpr = train_GPR(x_train, y_train)
gpr_train_pred, gpr_dev_pred = dev(gpr, x_train, y_train, x_dev, y_dev)

graph(train_dates, y_train, gpr_train_pred)
graph(dev_dates, y_dev, gpr_dev_pred)
```

```python
ElasticNet = train_ElasticNet(0.001, x_train, y_train)
EN_train_pred, EN_dev_pred = dev(ElasticNet, x_train, y_train, x_dev, y_dev)

graph(train_dates, y_train, EN_train_pred)
graph(dev_dates, y_dev, EN_dev_pred)
```

```python
def graph(dates, y_real, y_pred=None):    
    plt.plot(dates, y_real, color='tab:green')
    plt.plot(dates, y_pred, color='tab:red')
    plt.show()
```

```python
# citation: https://medium.com/@alexrachnog/neural-networks-for-algorithmic-trading-part-one-simple-time-series-forecasting-f992daa1045a

def time_window(x_train, y_train, window):
    X, Y = [], []
    for i in range(0, x_train.shape[0] - window):
        x_i = x_train[i:i + window]
        y_i = y_train[i + window]
        
        X.append(x_i)
        Y.append(y_i)
    return np.dstack(X).T, np.dstack(Y).T.reshape((-1))

df, ew_indret = process_data('industries/AERO.csv')
x_train, x_test, x_dev, y_dev, y_train, y_test = split_data(df, ew_indret)
```

```python
x_train, y_train = time_window(x_train, y_train, window=12)
x_dev, y_dev = time_window(np.asarray(x_dev), np.asarray(y_dev), window=12)

x_train = x_train.reshape(x_train.shape[0], -1)
x_dev = x_dev.reshape(x_dev.shape[0], -1)
```

```python
regr = train_LinearRegression(x_train, y_train)
regr_train_pred, regr_dev_pred = dev(regr, x_train, y_train, x_dev, y_dev)

graph(train_dates[12:], y_train, regr_train_pred)
graph(dev_dates[12:], y_dev, regr_dev_pred)
```

```python
reg = train_Ridge(400, x_train, y_train)
ridge_train_pred, ridge_dev_pred = dev(reg, x_train, y_train, x_dev, y_dev)

graph(train_dates[12:], y_train, ridge_train_pred)
graph(dev_dates[12:], y_dev, ridge_dev_pred)
```

```python
ElasticNet = train_ElasticNet(0.001, 3, x_train, y_train)
EN_train_pred, EN_dev_pred = dev(ElasticNet, x_train, y_train, x_dev, y_dev)

graph(train_dates[12:], y_train, EN_train_pred)
graph(dev_dates[12:], y_dev, EN_dev_pred)
```

```python
gpr = train_GPR(x_train, y_train)
gpr_train_pred, gpr_dev_pred = dev(gpr, x_train, y_train, x_dev, y_dev)

graph(train_dates[12:], y_train, gpr_train_pred)
graph(dev_dates[12:], y_dev, gpr_dev_pred)
```
