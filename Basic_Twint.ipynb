```
# Setup
import os
from os import path
import twint
import nest_asyncio
nest_asyncio.apply()
```

```
# User List
users = [
    # State media
    'IndyArabia', 'AlHadath', 'mbc1', 'Alarabiya',
    'saudi_gazette', 'arabnews', 'aawsat_News',
    'aawsat_eng', 'okaz_online', 'alwatansa', 'Khalejiatv', 'aleqtisadiah',
]
```

```
output_csv = "tweets.csv"

# Delete the CSV file if it exists
if path.exists(output_csv):
    os.remove(output_csv)

for i in users:
    # Configure
    c = twint.Config()
    c.Username = i
    c.Since = "2020-05-26 00:00:00"
    c.Until = "2020-05-27 00:00:00"
    c.Search = "covid OR coronavirus OR corona OR فيروس كورونا"
    c.Store_csv = True
    c.Output = output_csv
    # Run
    twint.run.Search(c)
```
